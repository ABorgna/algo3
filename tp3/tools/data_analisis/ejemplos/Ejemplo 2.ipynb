```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pylab as pl
import random
import numpy as np
```

```python
nodos = pl.frange(0,50,0.1)
clique = [x**5 for x in nodos]
a = [random.normalvariate(x**2,0.5) for x in nodos]
b = [random.normalvariate(x**2.5,0.5) for x in nodos]
```

# Como saber cuando experimentar
A la hora de sacar conclusiones es muy importante tener una buena idea de lo representativos que son nuestros datos, de lo contrario hay demasiada probabilidad de cometer errores en las conclusiones.

## Veamos un ejemplo extremo

Tenemos dos algoritmos que buscan de forma heurística cliques maximales en grafos y estamos midiendo el tamaño de las cliques. Para esto les pasamos grafos donde ya conocemos las cliques existentes de forma de poder ordenarlos en el eje x respeco de la clique maxima y en el eje y respecto de la clique encontrada por nuestro algoritmo.

```python
plt.clf()
df = pd.DataFrame({'Algoritmo A': a[0:6], 'Algoritmo B': b[0:6], 'Clique máxima': clique[0:6]})
df.plot(x='Clique máxima')
plt.ylabel('Clique encontrada')
plt.show()
```

```python
plt.clf()
df = pd.DataFrame({'Algoritmo A': a[0:10], 'Algoritmo B': b[0:10], 'Clique máxima': clique[0:10]})
df.plot(x='Clique máxima', logx=True)
plt.ylabel('Clique encontrada')
plt.show()
```

```python
plt.clf()
df = pd.DataFrame({'Algoritmo A': a[0:20], 'Algoritmo B': b[0:20], 'Clique máxima': clique[0:20]})
df.plot(x='Clique máxima')
plt.ylabel('Clique encontrada')
plt.show()
```

```python
plt.clf()
df = pd.DataFrame({'Algoritmo A': a, 'Algoritmo B': b, 'Clique máxima': clique})
df.plot(x='Clique máxima')
plt.ylabel('Clique encontrada')
plt.show()
```

## Cuando corremos experimentos, una sola instancia es valida ?
Cuando corremos experimentos hay muchas cosas que pueden afectar al resultado, es por esto que muchas veces no vasta con solo tomar una muestra para ver el resultado y es necesario correr varias veces el experimento.

Esto es muy común cuando se está midiendo tiempos y en los resultados de algoritmos no exactos. Como sabemos entonces cuantas iteraciónes hacer ?

```python
resultados = [random.normalvariate(50,50) for x in range(100000)]
resultados[0]
```

```python
plt.clf()
axis1 = plt.subplot(3,2,1)
axis2 = plt.subplot(3,2,2)
axis3 = plt.subplot(3,2,3)
axis4 = plt.subplot(3,2,4)
axis5 = plt.subplot(3,2,5)
axis6 = plt.subplot(3,2,6)

sns.distplot(resultados[0:2], ax=axis1)
sns.distplot(resultados[0:5], ax=axis2)
sns.distplot(resultados[0:10], ax=axis3)
sns.distplot(resultados[0:100], ax=axis4)
sns.distplot(resultados[0:10000], ax=axis5)
sns.distplot(resultados, ax=axis6)

sns.plt.show()
```

```python
[np.mean(resultados[0:i]) for i in [2,3,5,10,20,50,100,200,300,400,1000,2000,3000,6000,10000,30000,60000,100000]]
```

```python
df = pd.DataFrame({'mean': [np.mean(resultados[0:i]) for i in range(10000)], 'sample': range(10000)})
```

```python
plt.clf()
df.plot(x='sample')
plt.ylabel('mean')
plt.show()
```

```python
plt.clf()
df.plot(x='sample', logx=True)
plt.ylabel('mean')
plt.show()
```
