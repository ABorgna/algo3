```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Fijando dimensiones

```python
df = pd.DataFrame()
x = []
y = []
v = []
for i in range(50):
    for j in range(50):
        x.append(j)
        y.append(i)
        v.append((i+j)**2)
df['value'] = v
df['x'] = x
df['y'] = y
```

```python
plt.clf()
for i in [0,9,29,39,49]:
    df[df['y'] == i].plot(x='x',y='value')
plt.show()
```

## Ploteando en base a una función de las variables

```python
df = pd.DataFrame()
k = []
v = []
for i in range(50):
    for j in range(50):
        k.append(i+j)
        v.append((i+j)**2)
df['k'] = k
df['v'] = v

plt.clf()
ax = df.groupby('k').mean().plot()
plt.show()
```
